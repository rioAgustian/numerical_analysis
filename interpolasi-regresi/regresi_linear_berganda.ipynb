## Model

\begin{align}
y = a_0 + a_1x_1+a_2x_2
\end{align}

Cek:

https://simdos.unud.ac.id/uploads/file_pendidikan_1_dir/5f0221d2b0bb7ced1d61798fab7f4ad3.pdf

````python
# data
import numpy as np
x1 = np.array([0, 2, 2.5, 1, 4, 7])
x2 = np.array([0, 1, 2, 3, 4, 2])
y  = np.array([5, 10, 9, 0, 3, 27])
````

## Persamaan dalam bentuk matriks

\begin{align} 
        \begin{vmatrix}
        n & \Sigma x_{1,i} & \Sigma x_{2,i} \\
        \Sigma x_{1,i} & \Sigma x^2_{1,i} & \Sigma x_{1,i}x_{2,i}\\
        \Sigma x_{2,i} & \Sigma x_{1,i}x_{2,i} & \Sigma x^2_{2,i}
        \end{vmatrix} 
        \begin{vmatrix} a_0 \\ a_1 \\ a_2 \end{vmatrix} =
        \begin{vmatrix} \Sigma y_i \\ \Sigma x_{i,1}y_i \\ \Sigma x_{2,i}y_i \end{vmatrix}
    \end{align}

````python
# beberapa perhitungan
n = 6
sx1 = sum(x1)
sx2 = sum(x2)
sy  = sum(y) 
sx1_sqr = sum(x1**2)
sx2_sqr = sum(x2**2)
sy_sqr  = sum(y**2)
sx1y = sum(x1*y)
sx2y = sum(x2*y)
sx1x2 = sum(x1*x2)
````

````python
# matriks berikut dibuat untuk, mentukan nilai a0, a1, dan a2
A = [[n,   sx1,     sx2  ],
     [sx1, sx1_sqr, sx1x2],
     [sx2, sx1x2, sx2_sqr]]
B = [sy, sx1y, sx2y]

M  = [[n,  sx1,     sx2    ],
     [sx1, sx1_sqr, sx1x2  ],
     [sx2, sx1x2,   sx2_sqr]]
M1 = [[sy,   sx1,     sx2 ],
     [sx1y, sx1_sqr, sx1x2],
     [sx2y, sx1x2, sx2_sqr]]
M2 = [[n,   sy,  sx2    ],
     [sx1, sx1y, sx1x2  ],
     [sx2, sx2y, sx2_sqr]]
M3 = [[n,   sx1,    sy  ],
     [sx1, sx1_sqr, sx1y],
     [sx2, sx1x2,   sx2y]]
````

````python
# nilai a0, a1, dan a2
from numpy.linalg import det
a0 = det(M1)/det(M)
a1 = det(M2)/det(M)
a2 = det(M3)/det(M)
````

## Koefisien korelasi

\begin{align}
r = \sqrt{
  \frac{b_1\Sigma x_1y + b2 \Sigma x_2y}{\Sigma y^2}
}
\end{align}

````python
# koefisien korelasi
r = np.sqrt(
    ((a1*sx1y) + (a2*sx2y)) / sy_sqr
)
````

````python
# hasil 
# maksud "4" di samping f adalah banyaknya digit di belakang koma
# silahkan ubah untuk sesuai keperluan
a = "Persamaan: \ny = {0:0.4f} + {1:0.4f}x1 {2:0.4f}x2".format(a0,a1,a2)
b = "\n\nKoefisien korelasi: \nr = {0:0.4f}".format(r)
print(a + b)
````

````python
# data
import numpy as np
x1 = np.array([0, 2, 2.5, 1, 4, 7])
x2 = np.array([0, 1, 2, 3, 4, 2])
y  = np.array([5, 10, 9, 0, 3, 27])

# beberapa perhitungan
n = 6
sx1 = sum(x1)
sx2 = sum(x2)
sy  = sum(y) 
sx1_sqr = sum(x1**2)
sx2_sqr = sum(x2**2)
sy_sqr  = sum(y**2)
sx1y = sum(x1*y)
sx2y = sum(x2*y)
sx1x2 = sum(x1*x2)
# matriks berikut dibuat untuk, mentukan nilai a0, a1, dan a2
A = [[n,   sx1,     sx2  ],
     [sx1, sx1_sqr, sx1x2],
     [sx2, sx1x2, sx2_sqr]]
B = [sy, sx1y, sx2y]

M  = [[n,  sx1,     sx2    ],
     [sx1, sx1_sqr, sx1x2  ],
     [sx2, sx1x2,   sx2_sqr]]
M1 = [[sy,   sx1,     sx2 ],
     [sx1y, sx1_sqr, sx1x2],
     [sx2y, sx1x2, sx2_sqr]]
M2 = [[n,   sy,  sx2    ],
     [sx1, sx1y, sx1x2  ],
     [sx2, sx2y, sx2_sqr]]
M3 = [[n,   sx1,    sy  ],
     [sx1, sx1_sqr, sx1y],
     [sx2, sx1x2,   sx2y]]

# nilai a0, a1, dan a2
from numpy.linalg import det
a0 = det(M1)/det(M)
a1 = det(M2)/det(M)
a2 = det(M3)/det(M)

# koefisien korelasi
r = np.sqrt(((a1*sx1y) + (a2*sx2y)) / sy_sqr)

# hasil 
# maksud "4" di samping f adalah banyaknya digit di belakang koma
# silahkan ubah untuk sesuai keperluan
a = "Persamaan: \ny = {0:0.4f} + {1:0.4f}x1 {2:0.4f}x2".format(a0,a1,a2)
b = "\n\nKoefisien korelasi: \nr = {0:0.4f}".format(r)
print(a + b)
````
